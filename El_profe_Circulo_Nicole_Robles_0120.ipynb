## Formula para calcular Area
$A=/pi*r^2$

## Formula para calcular perimetro
$P=2*\pi*r$

## Datos de entrada
$r$ ?

## Diagrama de flujo en lucidchart
![Logo de Markdown](https://raw.githubusercontent.com/nicolerobles128/EPPL-Area-y-Perimetro-de-un-Circulo-mis-imagenes/refs/heads/main/DF%20Area%20Perimetro%20Circulo.PNG)

```python
# Calcular area y perimetro de un circulo
# Variables A,P, r
# Formula A = pi*r^  y P = 2 * pi * r
# Capturar el valor de r
import math
print("Calcular area y perimetro de un circulo")
r = float(input("Ingrese el valor de radio: "))
A = math.pi * r**2
P = 2 * math.pi * r
print("El area del circulo es: ", A)
print("El perimetro del circulo es: ", P)
print("Nicole Robles NL = 46")
```
